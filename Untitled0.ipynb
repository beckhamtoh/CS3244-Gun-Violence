**Upload the guns data**

```python
from google.colab import files
import pandas as pd

# Step 1: Upload the file
uploaded = files.upload()
```

**Read the CSV file**

```python
# Step 2: Load the uploaded CSV file
for filename in uploaded.keys():
    print(f"Loaded: {filename}")
    df = pd.read_csv(filename)

print(df.head())
```

**Data Info**


```python
print("df information")
df.info()

print("Describe df information")
df.describe(include='all')

print("Sum of NA")
df.isnull().sum()
```

```python
import matplotlib.pyplot as plt

df['Reason'].value_counts().head(10).plot(kind='bar')
plt.title('Top 10 Reasons for Incidents')
plt.show()
```

```python
import pandas as pd

# Convert Date column to datetime
df['Date'] = pd.to_datetime(df['Date'], dayfirst=True)

# Create Year-Month column
df['YearMonth'] = df['Date'].dt.to_period('M')

# Count incidents per month
monthly_counts = df.groupby('YearMonth').size()

# Convert to datetime for plotting
monthly_counts.index = monthly_counts.index.to_timestamp()

# Plot
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))
plt.plot(monthly_counts.index, monthly_counts.values, marker='o')
plt.title('Monthly Incident Trend')
plt.xlabel('Month')
plt.ylabel('Number of Incidents')
plt.grid(True)
plt.show()

```
